#### Desafio:
A resposta deve conter os valores da média, mediana, moda e desvio padrão da pontuação de crédito para cada estado do dataset.

### Os Dados:

```python
df = pd.read_csv('desafio1.csv')
df.head(5)
```

```python
# Descrição dos dados
df.describe()
```

### Solução

```python
# Agregando os resultados e calculando a Moda, Mediana, Média e Desvio Padrão.
df_to_json = df.groupby('estado_residencia').agg({
    'pontuacao_credito' : [
        pd.Series.mode,
        'median',
        'mean',
        'std'
    ] 
})

# Alterando o nome das colunas para adequar ao padrão do Json de envio.
df_to_json.columns = [
    'moda',
    'mediana',
    'media',
    'desvio_padrao'
]

# Criando o Json de Submissão
df_to_json.to_json('submission.json', orient='index')
```
